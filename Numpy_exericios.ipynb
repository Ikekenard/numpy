```python
# Seprando as laranjas e toranjas
laranja = dado[0:4999, 0:2]
toranja = dado[5000:10000, 0:2]
```

```python
# Sepratando diametro e Peso das Laranjas
diam_laranja = laranja[:, 0:1]
peso_laranja = laranja[:, 1:2]
```

```python
# Sepratando diametro e Peso das Toranjas
diam_toranja = toranja[:, 0:1]
peso_toranja = toranja[:, 1:2]
```

```python
# Gráfico Peso X Diametro das Laranjas
plt.plot(diam_laranja,peso_laranja)
```

```python
# Gráfico Peso X Diametro das Toranjas
plt.plot(diam_toranja,peso_toranja)
```

```python
x = diam_laranja
y = 0.52*x + 80
```

```python
plt.plot(diam_laranja,peso_laranja)
plt.plot(x,y)
```

```python
np.linalg.norm(diam_laranja - peso_laranja)
```

```python
Y = peso_laranja
X = diam_laranja
n = np.size(X)
```

```python
a = (n*np.sum(X*Y) - np.sum(X) * np.sum(Y)) / (n*np.sum(X**2) - np.sum(X)**2)
a
```

```python
# Coeficiente linear
b = np.mean(Y) - a* np.mean(X)
b
```

```python
y = a*X + b
```

```python
plt.plot(diam_laranja,peso_laranja)
plt.plot(x,y)
plt.plot(10,10*a+b,'*r')
```

```python
Y = peso_toranja
X = diam_toranja
n = np.size(X)
```

```python
a = (n*np.sum(X*Y) - np.sum(X) * np.sum(Y)) / (n*np.sum(X**2) - np.sum(X)**2)
a
```

```python
b = np.mean(Y) - a* np.mean(X)
b
```

```python
x = diam_toranja
```

```python
y = a*X + b
```
